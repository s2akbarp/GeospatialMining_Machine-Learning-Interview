```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a sample DataFrame
data = {
    'DEPTH': [1, 2, 3, 4, 5],
    'CTS': [10, 15, 20, 25, 30],
    'CTNI': [5, 10, 15, 20, 25],
    'CTCU': [2, 4, 6, 8, 10],
    'CTN': [3, 6, 9, 12, 15]
}
df = pd.DataFrame(data)

# Accessing a single column
cts = df['CTS']

# Accessing multiple columns
minerals = df[['CTS', 'CTNI', 'CTCU', 'CTN']]

# Accessing a single row by index
first_row = df.iloc[0]

# Accessing multiple rows by index range
first_three_rows = df.iloc[0:3]

# Accessing a specific cell
cell_value = df.iloc[0, 1]  # First row, second column

# Filtering rows based on a condition
filtered_df = df[df['CTS'] > 20]

# Adding a new column
df['new_column'] = df['CTS'] * 2

# Removing a column
df = df.drop(columns=['new_column'])

# Adding a new row
new_row = {'DEPTH': 6, 'CTS': 35, 'CTNI': 30, 'CTCU': 12, 'CTN': 18}
df = df.append(new_row, ignore_index=True)

# Removing a row
df = df.drop(0)

# Separate the DEPTH column
depth = df['DEPTH']
minerals = df.drop(columns=['DEPTH'])

# Standardize the minerals data
scaler = StandardScaler()
scaled_minerals = scaler.fit_transform(minerals)

# Convert the scaled data back to a DataFrame
minerals_scaled = pd.DataFrame(scaled_minerals, columns=minerals.columns)

# Combine the DEPTH column with the scaled minerals data
data_scaled = pd.concat([depth.reset_index(drop=True), minerals_scaled], axis=1)

# Check the scaled data
print(data_scaled.head())

# Line plot for each mineral based on DEPTH
plt.figure(figsize=(14, 8))
for column in data_scaled.columns[1:]:
    plt.plot(data_scaled['DEPTH'], data_scaled[column], marker='o', label=column)

plt.xlabel('Depth')
plt.ylabel('Standardized Mineral Value')
plt.title('Standardized Mineral Values Across Depths')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Histograms
data_scaled.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()

# Boxplots for checking outliers
plt.figure(figsize=(14, 8))
sns.boxplot(data=data_scaled.drop(columns=['DEPTH']))
plt.xticks(rotation=90)
plt.title('Boxplots of Standardized Mineral Values')
plt.show()
```
